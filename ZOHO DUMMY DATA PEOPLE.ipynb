```python
import random
import xlwt
```

```python
# create a list of Indian names
indian_names = {
    'first_name': ['Anjali', 'Amit', 'Ankush', 'Arjun', 'Deepak', 'Divya', 'Gaurav', 'Karan', 'Krishna', 'Manoj'],
    'last_name': ['Agarwal', 'Gupta', 'Jain', 'Kumar', 'Mehta', 'Sharma', 'Singh', 'Verma']
}
# create a list of Indian departments
departments = ['IT', 'HR', 'Marketing', 'Finance', 'Sales', 'Operations']
# create a list of Indian designations
designations = ['Manager', 'Associate', 'Consultant', 'Analyst', 'Officer', 'Sr. Manager']
#create a list of Indian Cities
cities = ['Delhi','Mumbai','Banglore','Hyderabad','Chennai','Kolkata']
#create a list of Indian states
states = ['Andhra Pradesh','Arunachal Pradesh','Assam','Bihar','Chhattisgarh','Goa','Gujarat','Haryana','Himachal Pradesh','Jharkhand','Karnatak']
#create a list of Indian employment types
employment_types = ['Full Time','Part Time','Remote','Temporary','Intern']
#create a list of Indian employee statuses
employee_statuses = ['Active','Inactive','Resigned','Terminated']
```

```python
book = xlwt.Workbook()
sheet1 = book.add_sheet("Sheet 1")
sheet1.write(0, 0, "Employee ID")
sheet1.write(0, 1, "First Name")
sheet1.write(0, 2, "Last Name")
sheet1.write(0, 3, "Nick Name")
sheet1.write(0, 4, "Email address")
sheet1.write(0, 5, "Department")
sheet1.write(0, 6, "Location")
sheet1.write(0, 7, "Designation")
sheet1.write(0, 8, "Zoho Role")
sheet1.write(0, 9, "Employment Type")
sheet1.write(0, 10, "Employee Status")
sheet1.write(0, 11, "Source of Hire")
sheet1.write(0, 12, "Date of Joining")
sheet1.write(0, 13, "Date of Birth")
sheet1.write(0, 14, "Gender")
sheet1.write(0, 15, "Marital Status")
sheet1.write(0, 16, "About Me")
sheet1.write(0, 17, "Ask me about/Expertise")
sheet1.write(0, 18, "Reporting Manager")
sheet1.write(0, 19, "UAN")
sheet1.write(0, 20, "PAN")
sheet1.write(0, 21, "Aadhaar")
sheet1.write(0, 22, "Work Phone Number")
sheet1.write(0, 23, "Extension")
sheet1.write(0, 24, "Seating Location")
sheet1.write(0, 25, "Tags")
sheet1.write(0, 26, "Address line 1(Present Address)")
sheet1.write(0, 27, "Address line 2(Present Address)")
sheet1.write(0, 28, "City(Present Address)")
sheet1.write(0, 29, "State(Present Address)")
sheet1.write(0, 30, "Country(Present Address)")
sheet1.write(0, 31, "Postal Code(Present Address)")
sheet1.write(0, 32, "Address line 1(Permanent Address)")
sheet1.write(0, 33, "Address line 2(Permanent Address)")
sheet1.write(0, 34, "City(Permanent Address)")
sheet1.write(0, 35, "State(Permanent Address)")
sheet1.write(0, 36, "Country(Permanent Address)")
sheet1.write(0, 37, "Postal Code(Permanent Address)")
sheet1.write(0, 38, "Personal Mobile Number")
sheet1.write(0, 39, "Personal Email Address")
sheet1.write(0, 40, "Date of Exit")

```

```python
for i in range(1, 101):
    # Generate a random employee ID
    employee_id = 'Emp_' + str(i)
    # randomly select first_name from list
    first_name = random.choice(indian_names['first_name'])
    # randomly select last_name from list
    last_name = random.choice(indian_names['last_name'])
    #randomly select a nickname
    nick_name = first_name[:3] + last_name[:3]
    #generate email address
    email = first_name.lower() + '.' + last_name.lower() + '@' + 'example.com'
    #randomly select department from the list
    department = random.choice(departments)
    #randomly select location from the list
    location = random.choice(cities)
    #randomly select designation from the list
    #randomly select Zoho Role
    zoho_role = random.choice(designations)
    #randomly select Employment Type
    employment_type = random.choice(employment_types)
    #randomly select Employee Status
    employee_status = random.choice(employee_statuses)
    #randomly select Source of Hire
    source_of_hire = random.choice(['Online', 'Referral', 'Walk-in'])
    #generate random date of joining 
    date_of_joining = '{:02d}-{:02d}-{:04d}'.format(random.randint(1,28),random.randint(1,12), random.randint(2010,2022))
    #generated date of birth in format DD-MM-YYYY
    date_of_birth = '{:02d}-{:02d}-{:04d}'.format(random.randint(1,28),random.randint(1,12), random.randint(1970,2000))
    #randomly select Gender
    gender = random.choice(['Male','Female'])
    #randomly select Marital Status
    marital_status = random.choice(['Married','Unmarried'])
    #Generate Random About me
    about_me = " "
    #Generate Random Ask me about/Expertise
    expertise = "I am expert in "+random.choice(['Java','Python','C++','C#','PHP','Javascript'])
    #Generate random Reporting Manager
    reporting_manager = random.choice(indian_names['first_name'])+" "+random.choice(indian_names['last_name'])
    #Generate random UAN
    uan = random.randint(111111111111, 999999999999)
    #Generate random PAN
    pan = 'ABCPEF'+str(random.randint(1111, 9999))
    #Generate random Aadhaar
    aadhaar = str(random.randint(111111111111, 999999999999))
    #Generate random Work phone number
    work_phone = str(random.randint(1111111111, 9999999999))
    #Generate random Extension
    extension = str(random.randint(1111, 9999))
    #Generate random Seating Location
    seating_location = 'Floor '+str(random.randint(1,10))+', Room '+str(random.randint(1,10))
    #Generate random Tags
    tags = ' '.join(random.choices(['IT', 'HR', 'Marketing', 'Finance', 'Sales', 'Operations'], k=3))
    #Generate Random Address line 1(Present Address)
    address_line1 = ' '.join(random.choices(['House', 'Apartment','Building'], k=2))+' '+str(random.randint(1,50))
    #Generate Random Address line 2(Present Address)
    address_line2 = ' '.join(random.choices(['Road', 'Street','Lane'], k=2))+' '+str(random.randint(1,50))
    #randomly select City(Present Address) from the list
    #randomly select City(Present Address) from the list
    city = random.choice(cities)
    #randomly select State(Present Address) from the list
    state = random.choice(states)
    #setting Country as India
    country = 'India'
    #generate random postal code
    postal_code = random.randint(11111, 99999)
    #Generate Random Address line 1(Permanent Address)
    address_line1_perm = ' '.join(random.choices(['House', 'Apartment','Building'], k=2))+' '+str(random.randint(1,50))
    #Generate Random Address line 2(Permanent Address)
    address_line2_perm = ' '.join(random.choices(['Road', 'Street','Lane'], k=2))+' '+str(random.randint(1,50))
    #randomly select City(Permanent Address) from the list
    city_perm = random.choice(cities)
    #randomly select State(Permanent Address) from the list
    state_perm = random.choice(states)
    #setting Country as India
    country_perm = 'India'
    #generate random postal code
    postal_code_perm = random.randint(11111, 99999)
    #Generate random Personal Mobile Number
    personal_mobile = str(random.randint(1111111111, 9999999999))
    #Generate random Personal Email Address
    personal_email = email.replace("example.com", "gmail.com")
    #Generate random Exit Date
    exit_date = '{:02d}-{:02d}-{:04d}'.format(random.randint(1,28),random.randint(1,12), random.randint(2023,2030))

    sheet1.write(i, 0, employee_id)
    sheet1.write(i, 1, first_name)
    sheet1.write(i, 2, last_name)
    sheet1.write(i, 3, nick_name)
    sheet1.write(i, 4, email)
    sheet1.write(i, 5, department)
    sheet1.write(i, 6, location)
    sheet1.write(i, 7, designations)
    sheet1.write(i, 8, zoho_role)
    sheet1.write(i, 9, employment_type)
    sheet1.write(i, 10, employee_status)
    sheet1.write(i, 11, source_of_hire)
    sheet1.write(i, 12, date_of_joining)
    sheet1.write(i, 13, date_of_birth)
    sheet1.write(i, 14, gender)
    sheet1.write(i, 15, marital_status)
    sheet1.write(i, 16, about_me)
    sheet1.write(i, 17, expertise)
    sheet1.write(i, 18, reporting_manager)
    sheet1.write(i, 19, uan)
    sheet1.write(i, 20, pan)
    sheet1.write(i, 21, aadhaar)
    sheet1.write(i, 22, work_phone)
    sheet1.write(i, 23, extension)
    sheet1.write(i, 24, seating_location)
    sheet1.write(i, 25, tags)
    sheet1.write(i, 26, address_line1)
    sheet1.write(i, 27, address_line2)
    sheet1.write(i, 28, city)
    sheet1.write(i, 29, state)
    sheet1.write(i, 30, country)
    sheet1.write(i, 31, postal_code)
    sheet1.write(i, 32, address_line1_perm)
    sheet1.write(i, 33, address_line2_perm)
    sheet1.write(i, 34, city_perm)
    sheet1.write(i, 35, state_perm)
    sheet1.write(i, 36, country_perm)
    sheet1.write(i, 37, postal_code_perm)
    sheet1.write(i, 38, personal_mobile)
    sheet1.write(i, 39, personal_email)
    sheet1.write(i, 40, exit_date)

book.save('employee_records.xls')





```
